#영화리뷰 데이터 불러오기

```python
lala_review = pd.read_csv('naver_preprocess.csv',encoding='utf-8')
lala_review.head()
```

```python
lala_review = lala_review.dropna()
```

#kosac 불러오기

```python
dic = pd.read_csv('polarity.csv')
```

```python
dic.head()
```

```python
dic = dic.loc[:,['ngram', 'NEG', 'POS']]
```

```python
dic.head()
```

```python
dic.info()
```

#전처리 된 Review를 리스트로

```python
review = lala_review.loc[:, ['body']]
```

```python
review = list(np.array(review['body'].tolist()))

print(review)
```

#ngram, pos, neg 리스트로

```python
ngram = list(np.array(dic['ngram'].tolist()))
pos = list(np.array(dic['POS'].tolist()))
neg = list(np.array(dic['NEG'].tolist()))
```

```python
print(ngram)
print(len(ngram))
```

```python
print(pos)
print(len(pos))
```

```python
print(neg)
print(len(neg))
```

#sentiment 예측 (y_pred)

```python
from konlpy.tag import Komoran

komoran = Komoran()

y_pred = []
y_true = []

for i in review:
  temp = komoran.pos(i)
  total = 0
  positive = 0
  negative = 0
  for j in temp:
    a = '/'.join(j)
    if a in ngram:
      index = ngram.index(a)
      negative += neg[index]
      positive += pos[index]
  total += positive - negative

  if total >= 5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
print(y_pred)
print(len(y_pred))
```

#실제 score 0, 1로 변환 (y_true)

```python
score = list(np.array(lala_review['score'].tolist()))

for i in score:
  if i>= 5:
    y_true.append(1)
  else:
    y_true.append(0)

print(y_true)
print(len(y_true))
```

#Classification report

```python
from sklearn.metrics import classification_report

target_names = ['negative', 'positive']
print(classification_report(y_true, y_pred, target_names=target_names))
```

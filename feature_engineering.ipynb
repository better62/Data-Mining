```python
# 상태(라벨)
y = []
for i in range(data_numpy.shape[0]):
  if data_numpy[i][0].endswith(('(자연사)', '(안락사)')):
    data_numpy[i][0] = 0
    y.append(0)

  else:
    data_numpy[i][0] = 1
    y.append(1)
y = np.array(y)   
print(y.shape)
```

```python
#공고번호
for i in range(data_numpy.shape[0]):
  location = data_numpy[i][1]
  data_numpy[i][1] = location[:2]
  
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
onehot_encoder = preprocessing.OneHotEncoder()

x_location = label_encoder.fit_transform(data_numpy[:,1])
x_location = x_location.reshape(-1,1) 
x_location = onehot_encoder.fit_transform(x_location).toarray()
#x_ㅣocation_recovery = np.argmax(x_location, axis=1).reshape(-1,1)
print(x_location.shape)
```

```python
#품종
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
onehot_encoder = preprocessing.OneHotEncoder()

x_species = label_encoder.fit_transform(data_numpy[:,3])
x_species = x_species.reshape(-1,1) 
x_species = onehot_encoder.fit_transform(x_species).toarray()
#x_ㅣocation_recovery = np.argmax(x_species, axis=1).reshape(-1,1)
print(x_species.shape)
```

```python
#성
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
onehot_encoder = preprocessing.OneHotEncoder()

x_sex = label_encoder.fit_transform(data_numpy[:,5])
x_sex = x_sex.reshape(-1,1) 
x_sex = onehot_encoder.fit_transform(x_sex).toarray()
#x_sex_recovery = np.argmax(x_sex, axis=1).reshape(-1,1)
print(x_sex.shape)
```
